```r
library(tidyverse)
library(cowplot)
library(ggsci)
```

```r
benchmarks = read.csv('tidy_benchmarks.csv', header = T, stringsAsFactors = FALSE)
benchmarks = benchmarks[-1]
```

```r
benchmarks$time[4] = 3.058538
benchmarks$time[12] = 4.891721
benchmarks$time[14] = 7.771492
benchmarks$time[26] = 2.864570
benchmarks$time[32] = 5.205193
benchmarks$method[15] = 'shapeit3'
benchmarks$method[18] = 'shapeit3'
benchmarks$method[19] = 'shapeit3'
runs = c('run1', 'run2', 'run4', 'run5', 'run6', 'run7', 'run8', 'run9', 'run10', 'run12', 'run13', 'run14', 'run15', 'run16', 'run17', 'run18', 'run19')
meth = c('eagle_combined', 'eagle_separate', 'shapeit', 'shapeit3', 'impute2', 'impute4', 'minimac')
benchmarks$modrun = factor(benchmarks$run, levels = runs) 
benchmarks$modmethod = factor(benchmarks$method, levels = meth)
```

```r
ggplot(benchmarks, aes(x = modrun, y = time, fill = modmethod)) +
    geom_bar(position = 'dodge', stat = 'identity', width = 0.75)+
    theme(axis.text.x = element_text(angle = 45, hjust = 1), aspect.ratio = 0.5, axis.title = element_text(size=12), 
         legend.text = element_text(size = 10))+
    xlab("") +
    ylab("Single-Threaded Runtime (Hours)")+
    guides(fill=guide_legend(title=""))
```

```r
list1 = c('run1', 'run2', 'run4', 'run5', 'run8', 'run17')
select(benchmarks, time, modmethod, modrun) %>% 
filter(modrun %in% list1) %>%
ggplot(aes(x = modrun, y = time, fill = modmethod)) +
    geom_bar(position = 'dodge', stat = 'identity', width = 0.75)+
    theme(axis.text.x = element_text(angle = 45, hjust = 1), aspect.ratio = 0.5, axis.title = element_text(size=12), 
         legend.text = element_text(size = 10))+
    xlab("") +
    ylab("Single-Threaded Runtime (Hours)")+
    guides(fill=guide_legend(title="")) +
    ggtitle("Phasing and Imputation Runtimes (List 1 Animals)")
```
